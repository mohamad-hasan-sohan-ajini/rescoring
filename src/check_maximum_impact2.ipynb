```python
import json
from pathlib import Path

import numpy as np
from tqdm import tqdm


# Example usage
path = Path("/home/aj/repo/rescoring/src/shard_checks")
json_files = sorted(path.glob("*.json"))
print(f"Found {len(json_files)} tar files")
```

```python
num_samples, asr_is_correct, rescoring_is_correct = 0, 0, 0
for json_file in tqdm(json_files):
    with open(json_file, "r") as f:
        shard_data = json.load(f)
    # Process shard_data as needed
    for sample in shard_data:
        gt = sample["transcription"]
        num_samples += 1
        if sample["hyps"][0] == gt:
            asr_is_correct += 1
        try:
            best_hyp_index = np.argmax(sample["avg_probs"]).item()
            if sample["hyps"][best_hyp_index] == gt:
                rescoring_is_correct += 1
        except ValueError:
            pass
```

```python
print(asr_is_correct / num_samples, rescoring_is_correct / num_samples)
```
